```python
# Read in city population data
import pandas

city_populations = pandas.read_csv("UN_cities_data.csv", sep='\t')
```

```python
# Compute annual population growth for years with real data. (I'm assuming 2015 and beyond are modeled)
city_populations["AverageAnnualGrowthRate"] = (city_populations["2010"] -
                                     city_populations["1950"])/(2010-1950)
```

```python
# create our data series for plotting
# I'm going to produce a scatter plot that uses lat/long for y/x coordinates with the points being the average annual
# population growth.  
x_Longitude = city_populations.Longitude
y_Latitude  = city_populations.Latitude
population_growth = city_populations["AverageAnnualGrowthRate"]
```

```python
%matplotlib inline
import matplotlib.pyplot as plt

fig=plt.figure(figsize=(16,8))
ax=fig.add_subplot(111)
ax.scatter(x_Longitude,y_Latitude,s=population_growth)
```
